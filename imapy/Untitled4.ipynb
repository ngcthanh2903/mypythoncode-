```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('72210768_1392808430883231_4870422059588517888_n.jpg',-123)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('Erosion', erosion)
cv2.imshow('Dilation', dilation)
cv2.waitKey(0)
plt.show()
plt.subplot(121),plt.imshow(img),plt.title('erosion')
plt.subplot(122),plt.imshow(img),plt.title('dilation')
```

```python
plt.subplot(121),plt.imshow(erosion),plt.title('erosion')
plt.subplot(122),plt.imshow(dilation),plt.title('dilation')
cv2.imshow('Erosion', erosion)
cv2.imshow('Dilation', dilation)
cv2.imshow('img',img)
cv2.waitKey()
print(erosion)
print(dilation)
```
